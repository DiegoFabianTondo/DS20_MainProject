```python
# Obtenemos la lista de Id de los tanques registrados para proceder a verificar los topes máximos en las transacciones
lista_id_tanques = set(tanques_df.id_tanque).intersection(transacciones_df.id_tanque)

# transacc_tanques_intersecc_df = pd.merge(transacciones_df, tanques_df, how='inner', on=['id_tanque'])
print(f'Al realizar la intersección de los dataframes obtenemos que el {transacciones_df.id_tanque.isin(tanques_df.id_tanque).sum()/len(transacciones_df)*100:.2f}'
      f'% de las transacciones son de tanques registrados.')
```

```python
# Se verifican las cantidades máximas en las transacciones en base a las capacidades de los tanques
def cantidad_vs_capacidad(transacciones_df, tanques_df, lista_id_tanques):
    cont = 0
    for Id in lista_id_tanques:
        cap_max = tanques_df[tanques_df['id_tanque'] == Id]['capacidad']
        cant = transacciones_df[transacciones_df['id_tanque'] == Id]['cantidad']
        cont += (cant  > cap_max.values[0]).sum()
    return cont

lista_id_tanques = set(tanques_df.id_tanque).intersection(transacciones_df.id_tanque)
cont = cantidad_vs_capacidad(transacciones_df, tanques_df, lista_id_tanques)

print(f'Número de transacciones con "cantidad" superior a la capacidad del tanque: {cont}')
```

```python
# Se imputa la capacidad del tanque para las cantidades de las transacciones que sobrepasan el valor de la capacidad del tanque
for Id in lista_id_tanques:
    cap_max = tanques_df[tanques_df['id_tanque'] == Id]['capacidad']
    cant = transacciones_df[transacciones_df['id_tanque'] == Id]['cantidad']
    transacciones_df.loc[(transacciones_df['id_tanque'] == Id) & (transacciones_df['cantidad'] > cap_max.values[0]), 'cantidad'] = cap_max.values[0]

# Verificamos que no quede ningún valor fuera de rango
cont = cantidad_vs_capacidad(transacciones_df, tanques_df, lista_id_tanques)
print(f'Número de transacciones con "cantidad" superior a la capacidad del tanque: {cont}')
```

### Feature Engineering

#### Obtención de la data periódica

Antes de obtener la serie temporal para los valores de demanda diarios de combustible, se realizará una imputación de outliers, ya que se observan valores que se salen del comportamiento normal de la demanda observada.

```python
data.isnull().sum()
```

```python
dicc_operations = {'producto': lambda x: statistics.mode(x),
                   'cantidad': 'sum',
                   'segmento': lambda x: statistics.mode(x),
                  }

data_period = data.set_index('fecha_hora').resample('120min').agg(dicc_operations)

# Obtenemos la variación diaria de la demanda de combustible
data_period['variacion_cantidad'] = data_period['cantidad'].diff()
data_period.dropna(inplace=True)
data_period.head()
```

```python
print('Segmentos de empresa que destaca en cada temporalidad:', *data_period.segmento.unique().tolist(), sep='\n- ')
```

**Valores perdidos o Missing values**

Verificamos que los días de la serie temporal esten completos. De no ser así se asigna un valor NaN al día faltante y se completaría con alguna técnica de imputación, o utilizando la función 'interpolate' de pandas: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.interpolate.html

**Valores atípicos o Outliers**

También se realizará un tratamiento de los valores atípicos (outliers).
Los datos de una serie de tiempo a menudo están contaminados con valores atípicos debido a la influencia de eventos inusuales y no repetitivos. La precisión del pronóstico en tales situaciones se reduce debido a un efecto de arrastre del valor atípico en el pronóstico puntual y un sesgo en las estimaciones de los parámetros del modelo.

```python
datetime_index = pd.date_range(data_period.index.min(), data_period.index.max(), freq='2H')
# Reindexamos y verificamos la presencia de algún valor nulo en la serie
data_period = data_period.reindex(datetime_index)
data_period.cantidad.isnull().sum()
```

Si observamos la longitud de los índices diarios creados, esta es igual a la longitud de la serie de tiempo diaria creada a partir de los datos.

### Tratamiento de outliers

```python
#===================================================
# Funciones para detección e imputación de outliers
#===================================================
def percent_change(serie):
    # Obtenemos los valores anteriores al valor actual y el valor actual, en base a los n periodos seleccionados
    previous_values = serie[:-1]
    last_value = serie[-1]

    percent_change = (last_value - np.mean(previous_values)) / np.mean(previous_values)
    return percent_change

def index_outliers(serie):
    diff_abs_valor_media = abs((serie) - (np.mean(serie)))

    # Calculate a mask for the differences that are > 3 standard deviations from zero
    mask = diff_abs_valor_media > 2 * np.std(serie)
    
    # Replace these values with the median accross the data
    index_out = serie[mask].index
    return index_out


# Cálculo del porcentaje de cambio respecto a la ventana de tiempo seleccionada
window = 12  # Equivalente a un día (2 hrs*12 = 24 hrs)
perc_change = data_period[['variacion_cantidad']].abs().rolling(window).apply(percent_change).fillna(method='backfill')

# Cálculo de los indices datetime en los cuales se detectaron outliers
index_out = perc_change.apply(index_outliers)

# Obtención de la mediana en base a una ventana de tiempo, para utilizarlo como valor de imputación de los outliers
medians = data_period['variacion_cantidad'].rolling(window).median().fillna(method='backfill')

# Imputación de la mediana a los outliers detectados
temp = data_period.copy() # se utiliza para graficar series comparativas del antes y después del tratamiento de utliers
data_period.loc[index_out.iloc[:,0].tolist(), 'variacion_cantidad'] = medians.loc[index_out.iloc[:,0].tolist()]
```

```python
fig, [ax1, ax2] = plt.subplots(2, 1, figsize=(20,6), sharex=True)
temp[['variacion_cantidad']].plot(ax=ax1)
ax1.set_title('Serie de tiempo de la Variación de Demanda antes del tratamiento de outliers')
data_period[['variacion_cantidad']].plot(ax=ax2, color='orange')
ax2.set_title('Serie de tiempo de la Variación de Demanda después del tratamiento de outliers')
plt.show()
```
